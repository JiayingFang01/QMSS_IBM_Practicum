```python
import pandas as pd
import matplotlib.pyplot as plt
Eo = pd.read_csv('/Users/zoezhuang/Desktop/presidential_executive_orders_final.csv')
print(Eo.head())
```

```python
# Filter out only Trump and Biden Executive Orders
Eo_filtered = Eo[(Eo["president"] == "Joseph R. Biden Jr.") | (Eo["president"] == "Donald Trump")]
Eo_filtered
```

```python
# drop duplicates
EO_cleaned = Eo_filtered.drop_duplicates()
# drop only the "significant" and "topics' column which contains all null values
Eo_cleaned = Eo_filtered.drop(columns=['significant','topics'])
# replace all missing values with 0
Eo_cleaned.fillna(0)
```

```python
Eo_cleaned['executive_order_notes'].head(10)
```

```python
# clean with executive_order_notes
# return no data if the value is missing
def clean_executive_order_notes(note):
    if pd.isna(note):
        return 'No Data'  
    
    # replace unwanted characters
    clean_note = note.replace('\r', ' ').replace('\n', ' ').replace('\t', ' ').strip()
    return clean_note

# aplly to dataframe 
Eo_cleaned['executive_order_notes'] = Eo_cleaned['executive_order_notes'].apply(clean_executive_order_notes)

Eo_cleaned.head()
```

```python
Eo_cleaned['agency_names'].head(10)
```

```python
# clean unwanted character of agency_names

def clean_names(item):
    an = str(['Executive Office of the President'])
    
  # drop '[' and ']' and "'"  
    an = an.replace('[', '').replace(']', '').replace("'", "").strip()
    
    return an

# apply it to the original column
Eo_cleaned['agency_names'] = Eo_cleaned.apply(clean_names, axis=1)


Eo_cleaned.head(25)

```

```python

import numpy as np
import re 

def clean_text(text):
    # delete "Start Printed Page" and the number after it
    # keep the beginning of text as 'Executive Order'
    text = re.sub(r"Start Printed Page \d+", "", text)
   # reduce unwanted character
    return text.replace('\n', ' ').strip()

# apply back 
Eo_cleaned['text'] = Eo_cleaned['text'].apply(clean_text)

# head of the text column
Eo_cleaned['text'].head(25)
```

```python
# Extract the executive order id from the 'text' column and create a new 'EO_id' column
# define a function to extract the executive order if from the text column
def extract_executive_order(text):
    match = re.search(r'(Executive Order\s+\d+)', text)
    if match:
        return match.group(1)
    return None

# apply the function and create a new 'EO_id' column
Eo_cleaned['EO_id'] = Eo_cleaned['text'].apply(extract_executive_order)
print(Eo_cleaned[['text', 'EO_id']].head())
```
